```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
pvA = 120000
rateA = 0.05
years = 5

fv5y = np.zeros(6)
fv5y [0] = pvA
for i in range (1,6):
    fv5y [i] = fv5y[i-1]*(1+0.05)
print("Przewidywana cena mieszkania wynosi: " + str("{:.2f}".format(fv5y[5])))
```

```python
freq = 12
rateB = 0.12
years = 5
fv = fv5y[5]

rateB /= freq  # konwersja stopy do okresu miesięcznego
nper = years * freq  # liczba wszystkich okresów
periods = np.arange(1,nper+1,dtype=int)

pmt = (fv*rateB)/(((1+rateB)**nper)-1)
print("Miesięczny wkład wynosi: " + str("{:.2f}".format(pmt)))
```

```python
deposit = pmt*(((1+rateB)**periods)-1)/rateB
```

```python
fv5yM = np.linspace(pvA,fv,num=60)
```

```python
plt.plot(fv5yM,label='cena mieszkania')
plt.plot(deposit,label='wartość lokaty')
plt.legend()
plt.xlabel('Liczba okresów')
plt.ylabel('Skumulowana wartość odsetek')
```
